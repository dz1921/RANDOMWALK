```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(12)
data_jum = np.random.lognormal(mean=0.875, sigma=0.3, size=1000 )
data_ang = np.random.lognormal(mean=0.5, sigma=0.5, size=1000)
data_wait = np.random.lognormal(mean=0.5, sigma=0.75, size=1000)
```

```python
plt.hist(data_jum, bins=100, density=True, color='blue', edgecolor='black', alpha=0.7)

plt.title('Jump or Step size')
plt.show()
```

```python
plt.hist(data_ang, bins=100, density=True, color='green', edgecolor='black', alpha=0.7)

plt.title('Turn angle')
plt.show()
```

```python
plt.hist(data_wait, bins=100, density=True, color='red', edgecolor='black', alpha=0.7)

plt.title('Wait time')
plt.show()
```

```python
# Plot histogram of the log-normal distribution
hist, bins, _ = plt.hist(data_jum, bins=100, density=True, color='blue', edgecolor='black', alpha=0.7)

# Calculate the width of each bin
bin_width = bins[1] - bins[0]

# Normalize the histogram
area = np.sum(hist * bin_width)
normalized_hist = hist / area

# Plot the normalized histogram
plt.bar(bins[:-1], normalized_hist, width=bin_width, color='blue', alpha=0.7, edgecolor='black')
plt.title('Normalized Histogram of Log-Normal Distribution')
plt.show()
```
