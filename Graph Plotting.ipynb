```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Distribution generation for HEALTHY
min_asph, max_asph = [0.0891, 0.891]
mean_asph = 0.420
std_asph = ((max_asph - min_asph)/4)
distribution_asph = np.random.normal(mean_asph, std_asph, 16)

min_str, max_str = [0.0694, 0.374] 
mean_str = 0.185
std_str = ((max_str - min_str)/4)
distribution_str = np.random.normal(mean_str, std_str, 16)

min_sin, max_sin = [0.165, 0.726]
mean_sin = 0.50
std_sin = ((max_sin - min_sin)/4)
distribution_sin = np.random.normal(mean_sin, std_sin, 16)

min_stp, max_stp = [1.04, 9.12] 
mean_stp = 2.20
std_stp = ((max_stp - min_stp)/4)
distribution_stp = np.random.normal(mean_stp, std_stp, 16)

min_ang, max_ang = [1.53, 2.28]
mean_ang = 1.82
std_ang = ((max_ang - min_ang)/4)
distribution_ang = np.random.normal(mean_ang, std_ang, 16)

min_vol, max_vol = [1.53, 2.28]
mean_vol = 1.82
std_vol = ((max_vol - min_vol)/4)
distribution_vol = np.random.normal(mean_vol, std_vol, 16)

def normalize(data):
    density = plt.hist(data, bins=30, density=True)[0]
    density /= density.sum()  # Normalize area to 1
    return density

norm_distribution_asph = normalize(distribution_asph)
norm_distribution_str = normalize(distribution_str)
norm_distribution_sin = normalize(distribution_sin)
norm_distribution_stp = normalize(distribution_stp)
norm_distribution_ang = normalize(distribution_ang)
norm_distribution_vol = normalize(distribution_vol)
```

```python
# Plotting
fig,ax=plt.subplots()
n=0.2
labels = ['Asphericity','Straightness','Sinuousity','Step length','Turn angle','Volume/time','']
labels.reverse()
plt.violinplot(norm_distribution_asph, positions=[6*n], vert=False, widths=0.1, showmeans=True)
plt.violinplot(norm_distribution_str, positions=[5*n], vert=False, widths=0.1, showmeans=True)
plt.violinplot(norm_distribution_sin, positions=[4*n], vert=False, widths=0.1, showmeans=True)
plt.violinplot(norm_distribution_stp, positions=[3*n], vert=False, widths=0.1, showmeans=True)
plt.violinplot(norm_distribution_ang, positions=[2*n], vert=False, widths=0.1, showmeans=True)
plt.violinplot(norm_distribution_vol, positions=[n], vert=False, widths=0.1, showmeans=True)


plt.xlim([-0.05,0.2])

# Add labels and title
plt.xlabel('Value')
""" ax.set_yticks([1,2,3,4,5,6]) """
ax.set_yticklabels(labels)
""" plt.xticks([1:6],labels) """
plt.title('Control cells (Normalised)')
plt.show()
```
